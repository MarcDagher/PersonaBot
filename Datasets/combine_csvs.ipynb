```python
import pandas as pd
import os
import re
```

```python
def replace_non_letters_with_space(text):
  text = re.sub(r"[^a-zA-Z]", " ", text)
  return text
```

```python
## Handles a folder like Cognitive/Abilities/file.csv
def create_combined_csv_from_a_multi_folder_folder(path_to_general_folder, category_type):
  # Go to the general folder "Cognitive Abilities and get the CA's categories"
  categories = os.listdir(path_to_general_folder)
  print(categories)

  # main_df will carry all the csvs
  main_df = pd.DataFrame()

  # Go from general folder to the categories
  for category in categories:
    path_to_csvs = path_to_general_folder + f"\{category}"
    csvs = os.listdir(path_to_csvs)

    # Go from categories to get the csvs
    for csv in csvs:
      path_to_csv = f"{path_to_csvs}\{csv}"
      
      # read_csv, add 2 column 'Ability' and 'Category, and concatenate the csv with the main_csv which containes all the abilities and their categories
      df = pd.read_csv(path_to_csv)
      ability_name = csv[:-4]
      df[category_type] = replace_non_letters_with_space(ability_name)
      df["Category"] = category
      main_df = pd.concat([df, main_df], ignore_index=True)

  print(f"main_df contains:")
  print(f"{len(list(main_df['Occupation'].unique()))} unique occupations")
  print(f"{len(list(main_df['Category'].unique()))} unique categories")
  print(f"{len(list(main_df[category_type].unique()))} unique {category_type}s")
  
  return main_df
```

```python
## Handles a folder like Abilities/file.csv
def create_combined_csv_from_a_non_multi_folder_folder(path_to_general_folder, category_name, category_type):
  
  # Go to the general folder "Cognitive Abilities and get the CA's categories"
  csvs = os.listdir(path_to_general_folder)
  print(csvs)

  # main_df will carry all the csvs
  main_df = pd.DataFrame()

  # Go from categories to get the csvs
  for csv in csvs:
    path_to_csv = f"{path_to_general_folder}\{csv}"
    
    # read_csv, add 2 column 'Ability' and 'Category, and concatenate the csv with the main_csv which containes all the abilities and their categories
    df = pd.read_csv(path_to_csv)
    ability_name = csv[:-4]
    df[category_type] = replace_non_letters_with_space(ability_name)
    df["Category"] = category_name
    main_df = pd.concat([df, main_df], ignore_index=True)

  print(f"main_df contains:")
  print(f"{len(list(main_df['Occupation'].unique()))} unique occupations")
  print(f"{len(list(main_df['Category'].unique()))} unique categories")
  print(f"{len(list(main_df[category_type].unique()))} unique {category_type}s")
  
  return main_df
```
